```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
xaxis = np.array([1,2,3,4,5])
xticks = xaxis

fig = plt.figure(figsize=[6,6])

synch = [None, 307.5584716796875, 281.2323455810547, 242.32772827148438, 233.5906219482422] 
asynch = list(reversed([None, 76.80303192138672, 88.30277252197266, 99.43669128417969, 106.90177917480469]))
CPU = [77.78, None, None, None, None]
GPU = [11.18, None, None, None, None]

synch = np.array(synch).astype(np.double) 
s_mask = np.isfinite(synch)
asynch = np.array(asynch).astype(np.double) 
a_mask = np.isfinite(asynch)
CPU = np.array(CPU).astype(np.double)
CPU_mask = np.isfinite(CPU)
GPU = np.array(GPU).astype(np.double)
GPU_mask = np.isfinite(GPU)

plt.plot(xaxis[s_mask],synch[s_mask],label='synch',linestyle='-',marker='o')
plt.plot(xaxis[a_mask],asynch[a_mask],label='asynch',linestyle='-',marker='o')
plt.plot(xaxis[CPU_mask],CPU[CPU_mask],label='CPU',linestyle='-',marker='o')
plt.plot(xaxis[GPU_mask],GPU[GPU_mask],label='GPU',linestyle='-',marker='o')

plt.xticks(xaxis,xticks)
plt.xlabel('Number of worker nodes')
plt.ylabel('Avg time taken per epoch')
plt.title('Comparison of training time')
plt.legend()
fig.savefig('training_time.png', dpi=fig.dpi)
```
